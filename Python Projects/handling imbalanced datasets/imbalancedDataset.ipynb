```python
#importing the dependancies
import numpy as np
import pandas as pd
```

```python
#loading dataset
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
credit_card_data.head()
```

```python
#distributuion of the two classes 
credit_card_data['Class'].value_counts()
```

This is a highly imbalanced dataset

0 --> legit transactions
1 --> fraud transactions

```python
#seperating the legit and fraud transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

UNDER-SAMPLING

```python
#build a sample dataset contaning similar distribution of legit and fraud transactions
#numner of fraud transaction - 492
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
#concatinate the two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
